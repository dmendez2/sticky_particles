```python
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np
import tifffile
```

```python
delta_val = "delta_25"
volume = tifffile.imread("data/multi_patch/lm_25_0/mxene_0_25/" + delta_val + ".tif")

# Show the middle z-slice
plt.imshow(volume[volume.shape[0] // 2], cmap='gray')
plt.title("Middle Z Slice")
plt.show()

delta_val = "delta_35"
volume = tifffile.imread("data/single_patch/lm_25_0/mxene_0_25/" + delta_val + ".tif")

# Show the middle z-slice
plt.imshow(volume[volume.shape[0] // 2], cmap='gray')
plt.title("Middle Z Slice")
plt.show()
```

```python

# Example: Load your 3D positions (shape: N x 3)
positions = np.load("data/single_patch/lm_25_0/mxene_0_25/delta_72.npy")  # or however you have them stored

# Extract x and y coordinates for a top-down view
x = positions[:, 0]
y = positions[:, 1]

plt.figure(figsize=(6, 6))
plt.scatter(x, y, s=2, c='black')  # s controls point size
plt.xlabel("x")
plt.ylabel("y")
plt.title("XY Projection of Particles")
plt.axis("equal")
plt.grid(True)
plt.show()
```

```python
z_low = 10.0
z_high = 12.0
slice_mask = (positions[:, 2] >= z_low) & (positions[:, 2] < z_high)
slice_positions = positions[slice_mask]

plt.scatter(slice_positions[:, 0], slice_positions[:, 1], s=2, c='black')
plt.title(f"XY Slice from z = {z_low} to {z_high}")
plt.axis("equal")
plt.show()
```

```python
file_name = "uniform_dist"
for i in range(0, 12):

    current_img = str(i)

    # Load the TIFF image
    if(i < 10):
        img = Image.open("data/" + file_name + "/slice_00" + current_img + ".tif")
    else: 
        img = Image.open("data/" + file_name + "/slice_0" + current_img + ".tif")

    # Display it
    plt.figure(figsize=(6, 6))
    plt.imshow(img, cmap='gray')  # use 'gray' for black/white
    plt.axis('off')  # Hide axes
    plt.title("Slice 010")
    plt.show()
```
